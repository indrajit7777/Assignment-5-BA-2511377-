### Task 6: CNN Concept Explanation

#### What is convolution?
In the context of CNNs, convolution is a mathematical operation that applies a filter (or kernel) to an input image to produce a feature map. The filter slides over the image, performing element-wise multiplication and summing the results for each position. This process helps the network detect patterns and features like edges, textures, or shapes in different parts of the image, regardless of their exact location.

#### Why is pooling used?
Pooling (specifically max pooling or average pooling) is used to reduce the spatial dimensions (width and height) of the feature maps. This reduction in size helps to:
1.  **Reduce computational cost:** Fewer parameters and computations are needed in subsequent layers.
2.  **Control overfitting:** By making the model less sensitive to the exact location of features.
3.  **Extract dominant features:** Max pooling, for example, selects the most active feature in a region, making the network more robust to small variations or distortions in the input.

#### Why is ReLU commonly used in CNNs?
ReLU (Rectified Linear Unit) is a popular activation function in CNNs because it helps to introduce non-linearity into the model. Without non-linearities, a deep network would behave like a single-layer network, unable to learn complex patterns. ReLU is computationally efficient (simply outputs the input if positive, otherwise zero), which speeds up training. It also helps to mitigate the vanishing gradient problem, allowing deeper networks to be trained more effectively.

#### Why are CNNs better than regular feed-forward networks for image data?
CNNs are superior for image data primarily due to their ability to leverage the spatial hierarchical structure of images:
1.  **Spatial Invariance:** Convolutional layers can detect features regardless of their position in the image. This is due to parameter sharing (the same filter is applied across the entire image) and pooling layers.
2.  **Hierarchical Feature Learning:** CNNs learn features at different levels of abstraction. Early layers detect basic features (edges), while deeper layers combine these to detect more complex features (shapes, objects).
3.  **Reduced Parameters:** Parameter sharing significantly reduces the number of trainable parameters compared to a fully connected feed-forward network, making CNNs more efficient and less prone to overfitting on high-dimensional image data.
4.  **Local Connectivity:** Neurons in convolutional layers are only connected to a small region of the input, mimicking the way biological visual systems work. This focuses on local patterns before combining them into global patterns.

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
import matplotlib.pyplot as plt
import numpy as np
import os
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import shutil # Import shutil for moving files

print("✅ TensorFlow version:", tf.__version__)

# Define the classes at the top for use in organizing files and model output
classes = ['normal', 'scratch', 'dent', 'stain']

# Define the root directory where the organized dataset will be placed
base_data_dir = 'dataset'

# --- FIX: Organize Dataset into Class Subdirectories ---
print("\n--- Organizing Dataset ---")
os.makedirs(base_data_dir, exist_ok=True)

# Create subdirectories for each class
for cls_name in classes:
    class_path = os.path.join(base_data_dir, cls_name)
    os.makedirs(class_path, exist_ok=True)

# DEBUG: Print contents of the current directory to verify files are seen
print(f"Contents of /content/ before file organization: {os.listdir('/content/')}")

# Move image files into their respective class subdirectories
moved_count = 0
# List files in the current working directory (which is /content/ in Colab)
for filename in os.listdir('/content/'): # Use '/content/' for current directory explicitly
    if filename.endswith(('.png', '.jpg', '.jpeg', '.bmp')):
        found_class = False
        for cls_name in classes:
            if filename.startswith(cls_name + '_'): # Check if filename starts with class name prefix
                source_path = os.path.join('/content/', filename) # Current directory explicitly
                destination_path = os.path.join(base_data_dir, cls_name, filename)
                # Check if the file already exists in the destination to prevent error if rerun
                if not os.path.exists(destination_path):
                    shutil.move(source_path, destination_path)
                    moved_count += 1
                found_class = True
                break
        # Optional: uncomment to warn about files that don't fit a class
        # if not found_class:
        #     print(f"Warning: Image '{filename}' does not match any defined class prefix and was not moved.")

print(f"Finished organizing dataset. Moved {moved_count} image files.")
# --- End of FIX ---

# Now, data_dir can correctly point to the newly organized 'dataset' folder
data_dir = base_data_dir

# =============================================
# TASK 1: Problem Type
# =============================================
print("This is an **Image Classification** problem.")

# =============================================
# TASK 2: Dataset Exploration
# =============================================
# data_dir = 'dataset'   # ←←← CHANGED TO MATCH YOUR ZIP - This is now set above
# classes = ['normal', 'scratch', 'dent', 'stain'] # This is now defined above

print("📊 Number of images per class:")
for cls in classes:
    path = os.path.join(data_dir, cls)
    num = len(os.listdir(path)) if os.path.exists(path) else 0
    print(f"   {cls}: {num} images")

# Show sample images
plt.figure(figsize=(12, 8))
for i, cls in enumerate(classes):
    # Ensure there are images in the directory before trying to access index 0
    class_image_files = os.listdir(os.path.join(data_dir, cls))
    if class_image_files: # Add check for empty list
        img_path = os.path.join(data_dir, cls, class_image_files[0]) # Use class_image_files
        img = load_img(img_path, target_size=(150, 150))
        plt.subplot(2, 2, i+1)
        plt.imshow(img)
        plt.title(cls.upper())
        plt.axis('off')
    else:
        print(f"Warning: No images found for class '{cls}'. Skipping sample display for this class.")
plt.suptitle('Sample Images from Each Class')
plt.show()

# =============================================
# TASK 3: Preprocessing
# =============================================
IMG_HEIGHT = 128
IMG_WIDTH = 128
BATCH_SIZE = 32

train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    horizontal_flip=True,
    validation_split=0.2
)

train_generator = train_datagen.flow_from_directory(
    data_dir, target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE, class_mode='categorical',
    subset='training', shuffle=True
)

validation_generator = train_datagen.flow_from_directory(
    data_dir, target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE, class_mode='categorical',
    subset='validation', shuffle=False
)

# =============================================
# TASK 4: Build Simple CNN
# =============================================
model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(IMG_HEIGHT, IMG_WIDTH, 3)),
    MaxPooling2D(2,2),
    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D(2,2),
    Conv2D(128, (3,3), activation='relu'),
    MaxPooling2D(2,2),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(len(classes), activation='softmax')
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()

# =============================================
# TASK 5: Train + Evaluate
# =============================================
# Only attempt to train if there are images found
if train_generator.samples > 0 and validation_generator.samples > 0:
    history = model.fit(train_generator, epochs=10, validation_data=validation_generator)

    # Save plots
    os.makedirs('results', exist_ok=True)
    os.makedirs('sample_predictions', exist_ok=True)

    # Accuracy & Loss curves
    plt.figure(figsize=(12,4))
    plt.subplot(1,2,1); plt.plot(history.history['accuracy'], label='Train'); plt.plot(history.history['val_accuracy'], label='Val'); plt.legend(); plt.title('Accuracy')
    plt.subplot(1,2,2); plt.plot(history.history['loss'], label='Train'); plt.plot(history.history['val_loss'], label='Val'); plt.legend(); plt.title('Loss')
    plt.savefig('results/accuracy_loss_curves.png')
    plt.show()

    # Confusion Matrix
    val_labels = validation_generator.classes
    val_pred_probs = model.predict(validation_generator)
    val_predictions = np.argmax(val_pred_probs, axis=1)

    cm = confusion_matrix(val_labels, val_predictions)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=classes, yticklabels=classes)
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title('Confusion Matrix')
    plt.savefig('results/confusion_matrix.png')
    plt.show()

    # Classification Report
    print('\nClassification Report:')
    print(classification_report(val_labels, val_predictions, target_names=classes))

    # Display some sample predictions
    plt.figure(figsize=(15, 10))
    for i in range(10): # Display 10 random images with predictions
        idx = np.random.randint(0, len(validation_generator.filenames))
        img_path = os.path.join(data_dir, validation_generator.filenames[idx])
        img = load_img(img_path, target_size=(IMG_HEIGHT, IMG_WIDTH))
        img_array = img_to_array(img)
        img_array = np.expand_dims(img_array, axis=0) / 255.0 # Preprocess image

        prediction = model.predict(img_array)
        predicted_class_idx = np.argmax(prediction)
        true_class_idx = validation_generator.classes[idx]

        plt.subplot(2, 5, i+1)
        plt.imshow(img)
        plt.title(f"True: {classes[true_class_idx]}\nPred: {classes[predicted_class_idx]}",
                  color='green' if predicted_class_idx == true_class_idx else 'red')
        plt.axis('off')
    plt.tight_layout()
    plt.savefig('sample_predictions/prediction_outputs.png')
    plt.show()

    print("🎉 All done! Check the folders: results/ and sample_predictions/")
else:
    print("❌ No images found for training or validation. Please ensure the dataset is correctly downloaded, unzipped, and organized.")
```
